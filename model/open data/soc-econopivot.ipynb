```python
import pandas as pd
```

```python
incomes = pd.read_excel('Incomes_BL.xlsx')
pop = pd.read_excel('pop_density.xlsx', sheet_name=5)
converter = pd.read_excel('zips_converter.xlsx')
```

```python
#Cleaning column names from imported data format
pop.columns = pop.iloc[0]
pop.drop(pop.index[0], inplace=True)
```

```python
pop.head(10)
```

```python
pop.columns = ['refnis', 'name_fr', 'name_nl', 'pop', 'size', 'pop_dens']
pop['refnis'] = pop.refnis.astype(int)
```

```python
pop.head()
```

```python
pop.dtypes
```

```python
converter.columns = ['zipcode', 'refnis', 'name_nl', 'name_fr']
```

```python
converter.head()
```

```python
convert_pop = pd.merge(converter, pop, on='refnis', how='left')
```

```python
convert_pop.drop(columns=['name_fr_y', 'name_nl_y', 'pop', 'size'], inplace=True)
```

```python
convert_pop.head()
```

```python
convert_pop[convert_pop.zipcode.duplicated(keep=False)]
```

```python
#dropping duplicated zipcode rows to maintain consistency
convert_pop = convert_pop.drop([3, 5, 110])
```

```python
convert_pop.rename(columns={'name_nl_x': 'name_nl', 'name_fr_x': 'name_fr'})
```

```python
incomes.columns
```

```python
incomes['income_per_capita'] = round(incomes['MS_TOT_NET_INC'] / incomes['MS_TOT_RESIDENTS'], 2)
incomes.head()
```

```python
province_mun_name = incomes[['CD_MUNTY_REFNIS', 'TX_MUNTY_DESCR_NL', 'TX_MUNTY_DESCR_FR', 'TX_PROV_DESCR_NL', 'TX_PROV_DESCR_FR', 'TX_RGN_DESCR_FR']].copy()
```

```python
province_mun_name.drop_duplicates(ignore_index=True, inplace=True)
```

```python
province_mun_name.rename(columns={'CD_MUNTY_REFNIS': 'refnis', 'TX_MUNTY_DESCR_NL': 'nl_name', 'TX_MUNTY_DESCR_FR': 'fr_name', 'TX_PROV_DESCR_NL': 'province_nl', 'TX_PROV_DESCR_FR': 'province_fr', 'TX_RGN_DESCR_FR': 'region'}, inplace=True)
```

```python
province_mun_name.fillna('Bruxelles', inplace=True)
```

```python
provincial_map = {
    'Bruxelles': 'Bruxelles',
    'Provincie West-Vlaanderen': 'West-Vlaanderen',
    'Provincie Vlaams-Brabant': 'Vlaams Brabant',    
    'Provincie Waals-Brabant': 'Brabant Wallon',
    'Provincie Antwerpen': 'Antwerpen',    
    'Provincie Oost-Vlaanderen': 'Oost-Vlaanderen',
    'Provincie Limburg': 'Limburg',
    'Provincie Luxemburg': 'Luxembourg',
    'Provincie Namen': 'Namur',
    'Provincie Luik': 'Liège',
    'Provincie Henegouwen': 'Hainaut'
}

province_mun_name['province'] = province_mun_name.province_nl.map(provincial_map)
```

```python
province_mun_name.province_nl.value_counts()
```

```python
province_mun_name.province.value_counts()
```

```python
province_mun_name.shape
```

```python
provinces_nl = ['Antwerpen', 'Vlaams Brabant', 'West-Vlaanderen', 'Oost-Vlaanderen', 'Limburg']
```

```python
def set_commune(row):
    provinces_nl = ['Antwerpen', 'Vlaams Brabant', 'West-Vlaanderen', 'Oost-Vlaanderen', 'Limburg']
    if row['province'] in provinces_nl:
        return row['nl_name']
    else:
        return row['fr_name']
```

```python
province_mun_name = province_mun_name.assign(commune=province_mun_name.apply(set_commune, axis=1))
```

```python
province_mun_name.drop(columns=['nl_name', 'fr_name', 'province_nl', 'province_fr', 'region'], inplace=True)
```

```python
province_mun_name.head()
```

```python
#mean incomes of all available years
incomes_17 = incomes.groupby(['CD_MUNTY_REFNIS']).income_per_capita.mean().round(2).to_frame()
incomes_17.sort_values(by='income_per_capita', ascending=False)
```

```python
#mean incomes of the last available year
incomes_1 = incomes.loc[incomes['CD_YEAR'] == 2022].groupby(['CD_MUNTY_REFNIS']).income_per_capita.mean().round(2).to_frame()
incomes_1.sort_values(by='income_per_capita', ascending=False)
```

```python
incomes_1 = incomes_1.reset_index()
```

```python
incomes_1.shape
```

```python
incomes_1.head()
```

```python
incomes_1.rename(columns={'CD_MUNTY_REFNIS': 'refnis', 'income_per_capita': 'income_1y'}, inplace=True)
```

```python
converter_full = pd.merge(convert_pop, incomes_1, on='refnis', how='left')
```

```python
incomes_17 = incomes_17.reset_index()
```

```python
incomes_17.rename(columns={'CD_MUNTY_REFNIS': 'refnis', 'income_per_capita': 'income_17y'}, inplace=True)
```

```python
incomes_17.head()
```

```python
converter_full = pd.merge(converter_full, incomes_17, on='refnis', how='left')
```

```python
converter_full['pop_dens'] = converter_full.pop_dens.astype(int)
```

```python
converter_full = pd.merge(converter_full, province_mun_name, on='refnis', how='left')
```

```python
converter_full.drop(columns=['name_nl_x', 'name_fr_x'], inplace=True)
```

```python
converter_full.rename(columns={'zipcode': 'zip_code'}, inplace=True)
```

```python
converter_full.to_csv('soceconvert.csv', index=False)
```
